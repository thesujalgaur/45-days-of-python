```python
import matplotlib.pyplot as plt
import numpy as np
import seaborn as set
import pandas as pd
import warnings 
```

```python
arr=np.random.randint(20000,40000,100)
arr
```

```python
##lineplot
arr.sort()
plt.plot(arr)
plt.show()
arr
```

```python
x=[2,3,4,5,6]
y=[4,5,6,7,9]
plt.plot(x,y,marker='D',ms=10,
         mfc="red",mec="green", #marker edge color # marke face color
         color="black",
         lw=3,ls=":", #line widh or style
         alpha=0.9,label="lineplot") #alpha teels transparency Range (0,1)
plt.title("lineplot")
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.legend()  # label or legend ko hamesha sath me likhne hota hai
plt.show()
```

```python
##Scatter plot works over numerical data 
plt.scatter(x,y,marker="d") #only show points called scatter plot
plt.show()
```

```python
x=["mp","up","rj","gj"]
y=[10,20,30,40]
colors=["red","green","blue","Yellow"]
hatch_patterns=["//","////","xx","oo"]
plt.bar(x,y,color=colors,hatch=hatch_patterns,
        label=x,width=0.3,edgecolor="black",alpha=0.7,linewidth=3)
#plt.barh(x,y) #for horizontal view of graph
plt.legend()
plt.grid()
plt.show()
```

```python
##HIstogram
#distribution of data in intervals (bins)
arr=np.random.randint(20000,40000,100)
plt.hist(arr,bins=20)
plt.show()
```

```python
#piechart-- percentage
x=["mp","up","rj"]
y=[10,20,30]
explodes=(0.1,0,0)
plt.pie(y,labels=x,autopct="%1.2f%%",shadow=True,explode=explodes)
plt.show() # explodes is used to make some different like the graph came out
```

```python
import warnings
warnings.filterwarnings("ignore")
##subplot
x=np.linspace(0,10,100)
plt.figure(figsize=(10,6))
plt.subplot(2,2,1) #2 rows 2 colnm ,1st graph
plt.plot(x,np.sin(x))
plt.title("sin")
plt.subplot(2,2,2) #2 rows 2 colnm ,1st graph
plt.plot(x,np.cos(x))
plt.title("cos")
plt.subplot(2,2,3) #2 rows 2 colnm ,1st graph
plt.plot(x,np.tan(x))
plt.title("tan")
plt.subplot(2,2,4) #2 rows 2 colnm ,1st graph
plt.plot(x,np.log(x))
plt.title("log")
plt.tight_layout()
plt.show()
```
